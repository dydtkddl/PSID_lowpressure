```python
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder

COREMOF = pd.read_csv("../2019-11-01-ASR-public_12020.csv")
GROUP = pd.read_csv("../Grouped_Results/Ar_results_V4(rerun_applied).csv").drop('Unnamed: 0', axis=1)

low = [0.01, 0.05, 0.1, 0.5]
high = [1, 5, 15]
T = 313

for l in low:
    for h in high:
        l_column = f"Ar_{T}_{l}"
        h_column = f"Ar_{T}_{h}"

        # GROUP 부분
        new_GROUP = GROUP.copy()[["MOF", l_column, h_column]].dropna()

        # COREMOF 부분
        COREMOF_ = COREMOF.copy()[[
            'filename', 'LCD', 'PLD', 'LFPD', 'cm3_g', 'ASA_m2_cm3', 'ASA_m2_g',
            'NASA_m2_cm3', 'NASA_m2_g', 'AV_VF', 'AV_cm3_g', 'NAV_cm3_g', 'Has_OMS'
        ]]

        # 병합
        MERGED = pd.merge(
            COREMOF_, new_GROUP,
            left_on="filename", right_on="MOF", how="inner"
        )

        # Has_OMS 인코딩
        encoder = LabelEncoder()
        MERGED["Has_OMS"] = encoder.fit_transform(MERGED["Has_OMS"])

        # 컬럼명 교체
        MERGED = MERGED.rename(columns={
            "MOF": "name",
            l_column: "Input",
            h_column: "Output"
        })

        # 파일명
        out_name = f"Ar_{T}K_Ar_{T}_{l}_to_Ar_{T}_{h}_dataset.csv"
        MERGED.to_csv(out_name, index=False)

        print(f"[Saved] {out_name} | shape={MERGED.shape}")
```

```python
import glob

# 데이터셋 찾기
datasets = glob.glob("Ar_313K/Ar_313K_Ar_313_*_dataset.csv")

trials = range(1, 6)  # 1~5
feature_sampler = [
    ("struct", "random_struct"),
    ("struct+input", "random_struct"),
    ("struct+input", "qt_then_rd"),
]
models = ["gbm", "cat", "rf"]

seed_base = 52  # 고정

cmds = []
for data in datasets:
    dataset_name = data.split("/")[-1].replace(".csv", "")
    for trial in trials:
        for feat, samp in feature_sampler:
            for model in models:
                d = data.replace("\\", "/")
                dd = d.replace("/", "_").replace(".csv", "")
                outdir = f"{dd}_{feat}_{samp}_{model}"
                cmd = (
                    f"python pipeline_single_trial.py "

                    f"--data {d} "
                    f"--outdir \"{outdir}\" "
                    f"--trial {trial} "
                    f"--seed-base {seed_base} "
                    f"--features {feat} "
                    f"--sampler {samp} "
                    f"--model {model}"
                )
                cmds.append(cmd)

# 저장
with open("command_list.txt", "w") as f:
    f.write("\n".join(cmds))

print(f"[Done] {len(cmds)} commands written to command_list.txt")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from pathlib import Path

# ----- 데이터 불러오기 -----
henry_result_AR273 = pd.read_csv("../Grouped_Results/07_henry_result_AR273.csv")[["name", "henry_coeff"]]
henry_result_AR293 = pd.read_csv("../Grouped_Results/07_henry_result_AR293.csv")[["name", "henry_coeff"]]
henry_result_AR313 = pd.read_csv("../Grouped_Results/07_henry_result_AR313.csv")[["name", "henry_coeff"]]
henry_result_HE273 = pd.read_csv("../Grouped_Results/07_henry_result_HE273.csv")[["name", "henry_coeff"]]
henry_result_HE293 = pd.read_csv("../Grouped_Results/07_henry_result_HE293.csv")[["name", "henry_coeff"]]
henry_result_HE313 = pd.read_csv("../Grouped_Results/07_henry_result_HE313.csv")[["name", "henry_coeff"]]

COREMOF = pd.read_csv("../2019-11-01-ASR-public_12020.csv")
encoder = LabelEncoder()
COREMOF = COREMOF[['filename', 'LCD', 'PLD', 'LFPD', 'cm3_g', 'ASA_m2_cm3', 'ASA_m2_g',
                   'NASA_m2_cm3', 'NASA_m2_g', 'AV_VF', 'AV_cm3_g', 'NAV_cm3_g','Has_OMS']]
COREMOF["Has_OMS"] = encoder.fit_transform(COREMOF["Has_OMS"])

AR = pd.read_csv("../Grouped_Results/Ar_results_V4(rerun_applied).csv")
HE = pd.read_csv("../Grouped_Results/He_results_V2(293Kadded).csv")

# ----- 매핑 -----
henry_dict = {
    ("Ar", 273): henry_result_AR273,
    ("Ar", 293): henry_result_AR293,
    ("Ar", 313): henry_result_AR313,
    ("He", 273): henry_result_HE273,
    ("He", 293): henry_result_HE293,
    ("He", 313): henry_result_HE313,
}
high_dict = {
    "Ar": AR,
    "He": HE
}

# ----- 파라미터 -----
HIGH_PRESSURE = [1, 5, 15]
TEMP = [273, 293, 313]
GAS = ["Ar", "He"]

OUTDIR = Path("./datasets_out")
OUTDIR.mkdir(exist_ok=True)

# ----- 루프 -----
for G in GAS:
    for T in TEMP:
        for H in HIGH_PRESSURE:
            # Henry
            df_henry = henry_dict[(G, T)]
            df_henry = df_henry.rename(columns={"henry_coeff": "Input"})  # Input = Henry coeff

            # High P uptake
            df_high = high_dict[G][["MOF", f"{G}_{T}_{H}"]].rename(columns={f"{G}_{T}_{H}": "Output"})

            # Join COREMOF + Henry + Uptake
            merged = pd.merge(COREMOF, df_henry, left_on="filename", right_on="name", how="inner")
            merged = pd.merge(merged, df_high, left_on="filename", right_on="MOF", how="inner")

            # drop extra cols
            merged = merged.drop(columns=["name", "MOF"])

            # Save
            fname = f"{G}_{T}K_{G}_{T}_Henry_to_{G}_{T}_{H}_dataset.csv"
            merged.to_csv(OUTDIR / fname, index=False)
            print(f"[SAVE] {fname} | rows={len(merged)}")
```
